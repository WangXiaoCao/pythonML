## 1.交叉验证score

```python
from sklearn import datasets, svm

digits = datasets.load_digits()
x = digits.data
y = digits.target

svc = svm.SVC(C=1, kernel='linear')
svc.fit(x[:-100], y[:-100]).score(x[-100:], y[-100:])
```

为了更好地度量精确度，我们可以使用交叉验证来进行训练与测试。

```python
import numpy as np

# 将特征集分成3份
x_folds = np.array_split(x, 3)
x_folds
```

```python
# 将标签集也分成3份
y_folds = np.array_split(y, 3)
y_folds
```

```python
# 分别对三份数据做交叉训练，然后提取三个score，这就叫做k-折交叉验证

scores = list()

for k in range(3):
    x_train = list(x_folds)
    x_test = x_train.pop(k)  # 提取索引为0的数据，作为测试集。原来的x_train数据里删除了这部分数据
    x_train = np.concatenate(x_train) # 合并剩下的2份数据，新城新的训练集
    y_train = list(y_folds)
    y_test = y_train.pop(k)
    y_train = np.concatenate(y_train)
    scores.append(svc.fit(x_train, y_train).score(x_test, y_test))
    
print scores
```

## 2.交叉验证生成器

### 2.1 实现CV的一种最简单的方法

```python
from sklearn import cross_validation
from sklearn import svm

# 随机地生成了一份交叉验证集
x_train, x_test, y_train, y_test = cross_validation.train_test_split(digits.data, digits.target,
                                                                    test_size=0.4, random_state=0)

clf = svm.SVC(kernel='linear', C=1).fit(x_train, y_train)
clf.score(x_test, y_test)
```

调用cross_validation.cross_val_score函数可以直接求交叉验证的score。

该函数接受某个estimator，数据集，对应的类标号，k-fold的数目，返回k-fold个score，对应每次的评价分数。

下面的例子会得到5个准确率。

```python
clf = svm.SVC(kernel='linear', C=1)
scores = cross_validation.cross_val_score(clf, digits.data, digits.target, cv=5)

scores
```

cross_val_score中的参数cv，既可以给定它一个整数，表示数据集被划分的份数（此时采取的是KFold或者StratifiedKFold策略，后面会说明）；也可以给定它一个CV迭代策略生成器，指定不同的CV方法。

下面的例子，首先通过cross_validation.ShuffleSplit生成一个CV迭代策略生成器cv，然后将cv以参数的形式传递到cross_val_score中。

```python
n_sample = digits.data.shape[0]

cv = cross_validation.ShuffleSplit(n_sample, n_iter=3, test_size=0.3, random_state=0)

cross_validation.cross_val_score(clf, digits.data, digits.target, cv=cv)
```

除了cross_val_score，sklearn中还提供一个cross_val_predict，它的功能就是返回每条样本作为CV中的测试集时，对应的模型对于该样本的预测结果。这就要求使用的CV策略能保证每一条样本都有机会作为测试数据，否则会报异常。

使用示例如下：

```python
predicted = cross_validation.cross_val_predict(clf, digits.data, digits.target, cv=10)
```
